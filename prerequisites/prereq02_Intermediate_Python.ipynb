```python
# 2D NumPy Arrays

import numpy as np
np_height = np.array([1.73, 1.68, 1.71, 1.89, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
meas = np.array([np_height, np_weight])
print("------2D NumPy Array------")
print(meas)
print("------use NumPy module function: nditer------")
for val in np.nditer(meas) : 
    print(val)
```

```python
# heads or tails: random walk
import numpy as np
np.random.seed(123)
tails = [0]
for x in range(10) :
    print("current position = " + str(x))
    coin = np.random.randint(0, 2)
    print("current position = " + str(x) + ", and the coin flip result is " + str(coin))
    tails.append(tails[x] + coin)
print(tails)
```

```python
# The next step
import numpy as np
np.random.seed(123)
# Initialize random_walk
random_walk = [0]

for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        # Replace below: use max to make sure step can't go below 0
        ## step = step - 1
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)


# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot random_walk
plt.plot(random_walk)

# Show the plot
plt.show()
```

```python
# 10,000 runs coin toss game
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)
final_tails = []
for x in range(10000) : # conduct this 10 throw experiment for 10,000 times
    tails = [0]
    for x in range(10) : # throw the coin 10 times
        coin = np.random.randint(0, 2)
        tails.append(tails[x] + coin)
    # indentation here specified that this last line is the top-level of the for loop
    final_tails.append(tails[-1])
plt.hist(final_tails, bins = 10)
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt

# initialize and populate all_walks
all_walks = []
for i in range(5) :
    random_walk = [0]
    for x in range(10) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)
print(all_walks)
# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw) # 10 lines → each one is a step, values for all walkers
plt.show()

# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw) 	# 5 lines → each one is a walker’s path over time (what you want)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Simulate random walk 500 times
all_walks = []
for i in range(500) :
    random_walk = [0] # random_walk[0] = starting position
    for x in range(100) : # random_walk[100] = final position after 100 steps
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)
# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
# The last row represents the final position (after 100 steps) for each of the 500 random walkers.
ends = np_aw_t[-1,:]

# Plot histogram of ends, display plot
# X-axis	Final height (number of steps climbed)
# Y-axis	Number of walkers who ended at that height
plt.hist(ends)
plt.show()

print("----- odds calculation-----")
# ends is a NumPy array of 500 final positions
# Count how many walkers reached at least step 60
successful_walks = np.sum(ends >= 60)

# Divide by total number of simulations (500)
odds = successful_walks / 500

print(f"Probability of reaching step ≥ 60: {odds:.2%}")
```
